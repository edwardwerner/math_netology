Загрузим датасет о ценах на дома в Бостоне. 

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.data.shape)
```

Каждый дом - это вектор в 13-мерном пространстве. 
Написать код для подсчета скалярного произведения двух первых домов из датасета.

```python
x1 = boston.data[0] # дом 1
x2 = boston.data[1] # дом 2
```

```python
x_y = x1.dot(x2)
x_y
```

Нарисовать два дома в 3-мерном пространстве (атрибуты выбрать самостоятельно)

```python
x_y = 0
for index in range(len(x1)):
    x_y += x1[index]*x2[index]
x_y
```

Найти косинус угла между векторами x1 и x2

```python
boston.feature_names
```

Найти расстояние между точками x1 и x2.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams["figure.figsize"] = (12,9)

crim=[x1[0], x2[0]]
age=[x1[6], x2[6]]
rad=[x1[8], x2[8]]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(crim, age, rad, s=80, c='b', marker='o')

ax.set_xlabel('CRIM')
ax.set_ylabel('AGE')
ax.set_zlabel('RAD')

plt.show()
```

Найти косинус угла между векторами x1 и x2

```python
import numpy as np
cos_x1_x2 = np.dot(x1,x2)/(np.linalg.norm(x1)*np.linalg.norm(x2))
cos_x1_x2
```

```python
import math

def norm(vec):
    res = 0
    for el in vec:
        res += el ** 2
    return math.sqrt(res)
```

```python
cos_x1_x2 = x_y / (norm(x1) * norm(x2))
cos_x1_x2
```

Найти расстояние между точками x1 и x2.

```python
dist_x1_x2 = np.linalg.norm(x1-x2)
dist_x1_x2
```

```python
div_x1_x2 = []
for index in range(len(x1)):
    div_x1_x2.append(x1[index] - x2[index])

dist_x1_x2 = norm(div_x1_x2)
dist_x1_x2
```

# Задание повышенного уровня.

Задан объект в пространстве 2D. 
Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).

```python
import numpy as np
import matplotlib.pyplot as plt

A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
    ]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
A_transf = [
    [1, 0, 300],
    [0, 1, 200],
    [0, 0, 1]
    ]
A_res = np.empty((0,3), int)
for point in A:
    new_point=np.dot(A_transf,np.transpose(point))
    A_res = np.append(A_res, [new_point], axis = 0)
x_new = np.array(A_res[:,0])
y_new = np.array(A_res[:,1])
plt.plot(x, y)
plt.plot(x_new, y_new)
plt.ylim([-200, 400])
plt.xlim([-200, 400])
```

Уменьшить объект в два раза.

```python
A_transf = [
    [0.5, 0, 0],
    [0, 0.5, 0],
    [0, 0, 1]
    ]
A_res = np.empty((0,3), int)
for point in A:
    new_point=np.dot(A_transf,np.transpose(point))
    A_res = np.append(A_res, [new_point], axis = 0)
x_new = np.array(A_res[:,0])
y_new = np.array(A_res[:,1])
plt.plot(x, y)
plt.plot(x_new, y_new)
plt.ylim([-200, 400])
plt.xlim([-200, 400])
```

Повернуть на 130 градусов.

```python
A_transf = [
    [math.cos(math.radians(130)), math.sin(math.radians(130)), 0],
    [-math.sin(math.radians(130)), math.cos(math.radians(130)), 0],
    [0, 0, 1]
    ]
A_res = np.empty((0,3), int)
for point in A:
    new_point=np.dot(A_transf,np.transpose(point))
    A_res = np.append(A_res, [new_point], axis = 0)
x_new = np.array(A_res[:,0])
y_new = np.array(A_res[:,1])
plt.plot(x, y)
plt.plot(x_new, y_new)
plt.ylim([-200, 400])
plt.xlim([-200, 400])
```
